<h2 style="padding:16px; color:white; background:#07D">Exercice D3 - Manipulation d'images</b></h2>

Une image numérique est définie comme une matrice de pixels où chaque pixel représente une couleur. Il existe trois grandes familles d'images selon la quantité d'information stockée dans chaque pixel :

- Les **images binaires** où chaque pixel ne contient qu'une valeur binaire 0 (= noir) ou 1 (= blanc)
- Les **images monochromes** où chaque pixel contient un scalaire correspondant à l'intensité lumineuse, qui peut être soit de type **`float`** (sur une échelle de 0 à 1), soit de type **`int`** (sur une échelle de 0 à 255)
- Les **images polychromes** où chaque pixel contient trois scalaires correspondant à l'intensité des trois couleurs primaires (***RGB*** en synthèse additive, ***CMY*** en synthèse soustractive) avec la même échelle que les images monochromes. Certains formats d'images (PNG, par exemple) utilisent une quatrième composante, appelée ***alpha*** et notée ***A***, qui permet de stocker une information de transparence

Dans l'écosystème ***Scientific Python***, on trouve les modules **`PIL`** (ou sa version plus récente, **`Pillow`**) ainsi que **`ndimage`** (fourni par le package **`scipy`**) qui ont été spécialement conçus pour mettre en oeuvre des traitements divers et variés sur des images (images de dimension 2 pour **PIL/Pillow**, images de dimensions arbitraires pour **ndimage**). Lorsqu'on souhaite faire du traitement d'images élaboré, ces modules sont évidemment à utiliser en priorité.

### 5 - Tables de correspondance

Une technique classique pour modifier le contenu d'une matrice à valeurs entières consiste à créer et utiliser une table de correspondance (***lookup table*** ou ***LUT*** en abrégé) qui permet d'associer une nouvelle valeur à chaque valeur initiale de la table. Dans le cas d'une image monochrome ou polychrome, les valeurs des pixels sont définies sur l'intervalle $[0, 255]$ donc la table doit posséder 256 éléments, chaque élément de la table étant lui-même défini sur l'intervalle $[0, 255]$.

Par exemple, on peut créer une table contenant 128x la valeur 0 et 128x la valeur 255 :

````python
table = np.repeat([0,255], 128); print(table)
````

Lorsqu'on applique une image monochrome à travers cette table de correspondance, elle est transformée en image binaire (tous les pixels inférieurs à 128 deviennent noirs et tous les pixels supérieurs à 127 deviennent blancs) :

````python
imshow(table[tigerG], cmap='gray');
````

Lorsqu'on applique la même table sur une une image polychrome, chacune des 3 composantes RGB est binarisée :

````python
imshow(table[tiger], cmap='gray');
````

Au lieu de créer les tables de correspondance directement, il est plus simple de définir une fonction continue de [0,1] dans [0,1] et de s'en servir pour créer la table contenant les 256 éléments. La fonction **`imshow_lut`** ci-dessous, permet de mettre ce principe en oeuvre :

````python
def imshow_lut(image, function):
  """create lookup table from 'function' and apply to 'image'"""
  ufunc = np.frompyfunc(function, 1, 1) # convert 'function' to ufunc
  ramp = np.linspace(0, 1, 256) # create ramp from 0 to 1 (256 items)
  table = (255*ufunc(ramp)).astype('u1') # create lookup table
  print(table[::5]) # show subsampling of lookup table
  imshow(table[image], cmap='gray') # apply lookup table and show image using 
````

La table et l'image précédente peuvent ainsi être générées avec une simple fonction **`binary`** qui place un seuil à 0.5 :

````python
binary = lambda t: 0 if t < 0.5 else 1
imshow_lut(tiger, binary)
````
