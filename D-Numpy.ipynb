<h2 style="padding:16px; color:white; background:#07D">Exercice D3 - Manipulation d'images</b></h2>

Une image numérique est définie comme une matrice de pixels où chaque pixel représente une couleur. Il existe trois grandes familles d'images selon la quantité d'information stockée dans chaque pixel :

- Les **images binaires** où chaque pixel ne contient qu'une valeur binaire 0 (= noir) ou 1 (= blanc)
- Les **images monochromes** où chaque pixel contient un scalaire correspondant à l'intensité lumineuse, soit en **`float`** (sur une échelle de 0 à 1), soit en **`int`** (sur une échelle de 0 à 255)
- Les **images polychromes** où chaque pixel contient trois scalaires correspondant à l'intensité des trois couleurs primaires (***RGB*** en synthèse additive, ***CMY*** en synthèse soustractive) avec la même échelle que les images monochromes. Certains formats d'images (PNG, par exemple) utilisent une quatrième composante, appelée *alpha* et notée ***A***, qui permet de stocker une information de transparence

L'objectif de cet exercice est d'utiliser les fonctionnalités du package **`numpy`** pour tester différentes manipulations sur des images qui auront été converties vers un conteneur de type **`array`**. Pour pouvoir charger, sauvegarder et visualiser ces images, nous allons utiliser certaines fonctions fournies par le package **`matplotlib`**. Ce package sera abordé en détail dans le chapitre 6 du cours, mais pour le moment, nous n'aurons besoin que de trois fonctions : **`imread(filename)`** pour charger l'image à partir d'un fichier, **`imshow(image)`** pour l'afficher dans le notebook, et **`imsave(filename, image)`** pour la sauvegarder dans un fichier.

````python
from matplotlib.pyplot import imread, imsave, imshow, style # import des fonctions
style.use('default') # utilisation du style par défaut pour la visualisation
````

````python
tiger = imread('IMG/tiger.jpg') # lecture de l'image sur disque et conversion en 'array'
imshow(tiger) # affichage de l'image avec les paramètres de visualisation par défaut
show("tiger.shape; tiger.dtype") # image polychrome = matrice 3D d'entiers : 480 x 640 x 3
````

````python
tigerG = imread('IMG/tigerG.jpg') # lecture de l'image sur disque et conversion en 'array'
imshow(tigerG, cmap='gray') # pour une image monochrome, il faut une palette de couleurs
show("tigerG.shape; tigerG.dtype") # image monochrome = matrice 2D d'entiers : 480 x 640
````

Une palette de couleur (***colormap***) permet d'associer une couleur à chaque valeur scalaire stockée dans les pixels d'une image monochrome. Le choix s'effectue en spécifiant le nom de la palette choise via le paramètre **`cmap`** de la fonction **`imshow`**. Parmi les palettes prédéfinies de **`matplotlib`**, on peut citer : 

***gray ● jet ● hot ● hsv ● brg ● rainbow ● prism ● flag ● seismic ● viridis ● Blues ● Greens ● Reds ● Purples ● Oranges***

La palette par défaut **`viridis`** a été choisie pour fournir un bon contraste, à la fois lorsqu'on la visualise en couleur sur un écran et lorsqu'on l'imprime sur papier en noir et blanc. Pour chaque palette, il existe une version inversée, obtenue en rajoutant le suffixe **`_r`** au nom de la palette
