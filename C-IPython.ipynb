> **Note :** La fonction **`latin(seq)`** utilise la fonction **`randrange`** pour créer les permutations aléatoires de la matrice, donc chaque exécution va donner un résultat différent, ce qui rend les tests unitaires systématiques impossibles. La solution consiste à forcer le générateur aléatoire à fournir exactement la même séquence de valeurs à chaque appel, grâce à la fonction **`seed`** du module **`random`**

````python
# gcd(p,q)
````

````python
def gcd(p:'int > 0', q:'int > 0') -> 'int > 0':
  """return the Greatest Common Divisor for p and q
  >>> gcd(5,2), gcd(10,6), gcd(9,12), gcd(8,36), gcd(50,15)
  (1, 2, 3, 4, 5)
  >>> gcd(3,0), gcd(6,-2), gcd(1.5,3)
  Traceback (most recent call last):
  ...
  AssertionError: 3 and 0 must be strictly positive integers
  """
  assert isinstance(p, int) and isinstance(q, int) and p > 0 and q > 0, \
    f"{p} and {q} must be strictly positive integers"
  while q: p, q = q, p % q # compute GCD by using iterative version of Euclids' algorithm
  return p
````

````python
# lcm(p,q)
````

````python
def lcm(p: 'int > 0', q: 'int > 0') -> 'int > 0':
  """return the Least Common Multiple for p and q
  >>> lcm(5,2), lcm(10,6), lcm(9,12), lcm(8,36), lcm(50,15)
  (10, 30, 36, 72, 150)
  >>> lcm(3,0), lcm(6,-2), lcm(1.5,3)
  Traceback (most recent call last):
  ...
  AssertionError: 3 and 0 must be strictly positive integers
  """
  return p*q // gcd(p,q) # use GCD to get LCM
````

````python
# gcd_lcm(p,q)
````

````python
def gcd_lcm(n: 'int > 0') -> 'str (CSV format)':
  """return the GCD/LCM table of size 'n' x 'n', as a CSV string
  >>> gcd_lcm(20) # doctest: +ELLIPSIS
  '  1,   1, ... 380,  20\\n'
  """
  csv, colsep, rowsep = '', ', ', '\n' # set col and row separator
  size = len(str(n*n)) # define max nb of chars for table cells
  for p in range(1, n+1): # loop over table rows
    for q in range(1, n+1): # loop over table cols
      sep = colsep if q < n else rowsep # set separator for current cell
      val = gcd(p,q) if p < q else lcm(p,q) # set value for current cell
      csv += f"{val:{size}}{sep}" # append value and separator to CSV string
  return csv
````

````python
def prime_eratos(n: 'int > 1') -> 'list of prime numbers':
  """compute all prime numbers from 2 to 'n', using the sieve of Eratosthenes
  >>> print(prime_eratos(1000)) # doctest: +ELLIPSIS
  [2, 3, 5, 7, ... 983, 991, 997]
  >>> lcm(3,0), lcm(6,-2), lcm(1.5,3)
  Traceback (most recent call last):
  ...
  AssertionError: 3 and 0 must be strictly positive integers
  """
  assert isinstance(n, int) and n > 1, f"{n} must be an integer > 1"
  primes, notprimes = [2], set() # list for primes, set for non-primes
  for p in range(3, n+1, 2): # loop over all odd numbers from 3 to 'n'
    if p in notprimes: continue # switch to next value
    primes.append(p) # 'p' is prime so append it to 'primes'
    # all odd multiples of 'p' are non-primes, so add them to 'notprimes'
    for q in range(p*p, n+1, 2*p): notprimes.add(q)
  return primes # return final list with all primes up to 'n'
````

````python
def latin(seq: 'iterable') -> '2D list of int':
  """return a random latin square from the provided sequence
  >>> from random import seed; seed(0); latin([1,2,3,4])
  [[4, 3, 1, 2], [3, 2, 4, 1], [1, 4, 2, 3], [2, 1, 3, 4]]
  >>> latin(0)
  Traceback (most recent call last):
  ...
  AssertionError: 0 must be iterable
  """
  from random import randrange as rr # local import of an external function
  assert hasattr(seq, '__iter__'), f"{seq} must be iterable"
  seq = list(seq); n = len(seq) # convert 'seq' to list and get its length
  mat = [seq[p:] + seq[:p] for p in range(n)] # rotate 'seq' by 'p' steps on row 'p'
  for loop in range(n*n): # loop 'n*n' times to get pretty good shuffling
    p, q = rr(n), rr(n) # randomly select two row indices
    for r in range(n): # exchange row 'p' with row 'q'
      mat[p][r], mat[q][r] = mat[q][r], mat[p][r]
    p, q = rr(n), rr(n) # randomly select two col indices
    for r in range(n): # exchange col 'p' with col 'q'
      mat[r][p], mat[r][q] = mat[r][q], mat[r][p]
  return mat
````

````python
def show_grid(mat:'2D list') -> None:
  """display provided matrix 'mat' as a grid, using '+|-' for framing
  >>> show_grid([[1,2,3],[4,5,6],[7,8,9]])
  +---+---+---+
  | 1 | 2 | 3 |
  +---+---+---+
  | 4 | 5 | 6 |
  +---+---+---+
  | 7 | 8 | 9 |
  +---+---+---+
  >>> show_grid([1,2,3,4,5,6,7,8,9])
  Traceback (most recent call last):
  ...
  TypeError: object of type 'int' has no len()
  """
  rows, cols = len(mat), len(mat[0]) # get nb of rows and nb of cols for matrix
  mat = [[str(cell) for cell in row] for row in mat] # convert all matrix cells to strings
  size = max(max(len(cell) for cell in row) for row in mat) # get max string size for all cells
  mat = [[f"{cell:^{size}}" for cell in row] for row in mat] # center all strings on 'size' chars
  colsep, rowsep = ' | ', ' |\n' + cols * ('+--' + size*'-') + '+\n| ' # set col and row separators
  grid = rowsep + rowsep.join(colsep.join(row) for row in mat) + rowsep # convert to multiline string
  print(grid[3:-3]) # remove ' |\n' and '\n| ' at string boundaries
````

Au lieu d'afficher tout le chemin d'un seul coup, on peut utiliser les fonctions d'animation disponibles en SVG pour animer le marcheur sous la forme d'un carré vert qui se déplace entre la case d'entrée et celle de sortie. Pour cela, on utilise la balise SVG **`<animateMotion>`** qui permet d'interpoler la position du marcheur, en spécifiant une série de positions successives :

````python
def show_walk(maze, walk, scale=8, speed=1):
  """display provided maze and walk as an animated SVG image, using 'scale' and 'speed' factors"""
  rows, cols = len(maze), len(maze[0]) # get nb of rows and nb of cols for maze
  view = f"width='{scale*cols}' viewBox='0 0 {cols} {rows}'" # define view for SVG
  head, tail = f"<svg {view}>", '</svg>' # define head and tail of SVG code
  colors, time = {'A':'#000', 'B':'#FFF'}, len(walk)/speed/10 # define colors and animation time
  svg = [f"<rect x='{c}' y='{r}' width='1' height='1' fill='{colors[maze[r][c]]}'/>"
    for r in range(rows) for c in range(cols)] # create list of <rect> tags for all maze cells
  path = ' '.join(f"L{x},{y}" for x,y in (step.split() for step in walk))[1:] # generate animation path
  svg += [f"""<rect width='1' height='1' fill='#0F0'>
    <animateMotion dur='{time}' repeatCount='indefinite' path='M{path}'/></rect>"""] # generate animation
  svg = '\n'.join([head, *svg, tail]) # merge SVG head, body and tail
  #cutcut(svg) # print head and tail lines of SVG code
  #dp.display(dp.SVG(svg)) # display SVG image
  dp.display(dp.HTML(f"<center>{dp.SVG(svg).data}</center>")) # display centered SVG image
````

L'autre possibilité pour l'animation est de dessiner le chemin de manière incrémentale, case par case. Pour cela, on utilise la balise SVG **`animate`** qui permet de changer une propriété au cours du temps. L'idée est de commencer à dessiner le chemin comme des cases blanches qui vont être coloriées en vert de manière séquentielle :

````python
def show_walk(maze, walk, scale=8, speed=1):
  """display provided maze and walk as an animated SVG image, using 'scale' and 'speed' factors"""
  rows, cols = len(maze), len(maze[0]) # get nb of rows and nb of cols for maze
  view = f"width='{scale*cols}' viewBox='0 0 {cols} {rows}'" # define view for SVG
  head, tail = f"<svg {view}>", '</svg>' # define head and tail of SVG code
  colors, time = {'A':'#000', 'B':'#FFF'}, 0.1/speed # define colors and animation time
  svg = [f"<rect x='{c}' y='{r}' width='1' height='1' fill='{colors[maze[r][c]]}'/>"
    for r in range(rows) for c in range(cols)] # create list of <rect> tags for all maze cells
  svg += [f"""<rect x='{x}' y='{y}' width='1' height='1' fill='#FFF'>
    <animate attributeName='fill' from='#0F0' to='#0F0' begin='{time*n}s' dur='indefinite'/></rect>"""
    for n,(x,y) in enumerate(step.split() for step in walk)] # create <rect> tags for all walk steps
  svg = '\n'.join([head, *svg, tail]) # merge SVG head, body and tail
  #cutcut(svg) # print head and tail lines of SVG code
  dp.display(dp.SVG(svg)) # display SVG image
  #dp.display(dp.HTML(f"<center>{dp.SVG(svg).data}</center>")) # display centered SVG image
````
