Les [**équations de Maxwell**](https://fr.wikipedia.org/wiki/%C3%89quations_de_Maxwell), publiées par [**James Clerk Maxwell**](https://fr.wikipedia.org/wiki/James_Clerk_Maxwell) en 1865, puis simplifiées par [**Oliver Heaviside**](https://fr.wikipedia.org/wiki/Oliver_Heaviside) en 1884, synthétisent les lois fondamentales de l'électromagnétisme, en reliant le champ électrique $\vec{\bf E}$ et le champ magnétique $\vec{\bf B}$ par un ensemble de 4 équations différentielles, exprimées dans les 4 dimensions de l'espace-temps $(x,y,z,t)$ :

$$\begin{align}
\nabla \times \vec{\bf B} & = \mu_0\, \sigma\, \vec{\bf E} \,+\, \mu_0\, \varepsilon_0\, \frac{\partial\vec{\bf E}}{\partial t} &&
\textrm{(Equation de Maxwell-Ampère)} \\
\nabla \times \vec{\bf E} & = -\, \frac{\partial\vec{\bf B}}{\partial t}  &&
\textrm{(Equation de Maxwell-Faraday)} \\
\nabla \cdot \vec{\bf E} & = \frac{\rho}{\varepsilon_0} &&
\textrm{(Equation de Maxwell-Gauss)} \\
\nabla \cdot \vec{\bf B} & = 0 &&
\textrm{(Equation de Maxwell-Thomson)}
\end{align}$$

où $\sigma$ représente la conductivité électrique, $\rho$ la densité de charge, $\varepsilon_0$ la permittivité
électrique du vide et $\mu_0$ la perméabilité magnétique du vide
