Une image numérique est définie comme une matrice de pixels où chaque pixel représente une couleur. Il existe trois grandes familles d'images selon la quantité d'information stockée dans chaque pixel :

- Les **images binaires** où chaque pixel ne contient qu'une valeur binaire 0 (= noir) ou 1 (= blanc)
- Les **images monochromes** où chaque pixel contient un scalaire correspondant à l'intensité lumineuse, qui peut être soit de type **`float`** (sur une échelle de 0 à 1), soit de type **`int`** (sur une échelle de 0 à 255)

Certains formats d'images (PNG, par exemple) utilisent une quatrième composante, appelée ***alpha*** et notée ***A***, qui permet de stocker une information de transparence

### 5 - Tables de correspondance

Une technique classique pour modifier le contenu d'une matrice à valeurs entières consiste à créer et utiliser une table de correspondance (***lookup table*** ou ***LUT*** en abrégé) qui permet d'associer une nouvelle valeur à chaque valeur initiale de la table. Dans le cas d'une image monochrome ou polychrome, les valeurs des pixels sont définies sur l'intervalle $[0, 255]$ donc la table doit posséder 256 éléments, chaque élément de la table étant lui-même défini sur l'intervalle $[0, 255]$.

Par exemple, on peut créer une table contenant 128x la valeur 0 et 128x la valeur 255 :

````python
table = np.repeat([0,255], 128); print(table)
````

Lorsqu'on applique une image monochrome à travers cette table de correspondance, elle est transformée en image binaire (tous les pixels inférieurs à 128 deviennent noirs et tous les pixels supérieurs à 127 deviennent blancs) :

````python
view(table[tigerG]);
````

Lorsqu'on applique la même table sur une une image polychrome, chacune des 3 composantes RGB est binarisée :

````python
view(table[tiger]);
````

Au lieu de créer les tables de correspondance directement, il est plus simple de définir une fonction continue de [0,1] dans [0,1] et de s'en servir pour créer la table contenant les 256 éléments. La fonction **`view_lut`** ci-dessous, permet de mettre ce principe en oeuvre :

````python
def view_lut(image, function):
  """create lookup table from 'function' and apply to 'image'"""
  ufunc = np.frompyfunc(function, 1, 1) # convert 'function' to ufunc
  ramp = np.linspace(0, 1, 256) # create ramp from 0 to 1 (256 items)
  table = (255*ufunc(ramp)).astype('u1') # create lookup table
  print(table[::5]) # show subsampling of lookup table (1 sample out of 5)
  view(table[image]) # apply lookup table and show image
````

La table et l'image précédente peuvent ainsi être générées avec une simple fonction **`binary`** qui place un seuil à 0.5 :

````python
binary = lambda t: 0 if t < 0.5 else 1
view_lut(tiger, binary)
````

<h2 style="padding:16px; color:#FFF; background:#07D">Exercice D4 - Solveur de Sudoku</h2>

Le jeu du [**Sudoku**](https://fr.wikipedia.org/wiki/Sudoku) est un jeu de logique faisant partie de la famille des [**Constraint Satisfaction Problems**](https://en.wikipedia.org/wiki/Constraint_satisfaction_problem) (***CSP***), tout comme l'énigme du zèbre présentée dans l'exercice B9. De ce fait, le module [**python-constraint**](https://labix.org/python-constraint) mentionné dans l'exercice B9 est donc parfaitement adapté à la création d'un solveur de grilles de Sudoku, qui sera à la fois très efficace en temps de résolution, mais surtout très compact (moins d'une quinzaine de lignes de code sont nécessaires).

Mais l'objectif de cet exercice est de réaliser une implémentation alternative d'un solveur de Sudoku qui utilisera uniquement la bibliothèque **numpy**. Le principe général pour l'algorithme à implémenter est extrêmement simple et se base sur une unique structure de données : une variable **`guess`** correspondant à une matrice booléenne 3D de taille ***9x9x9***. Les 2 premières dimensions de cette matrice correspondent aux lignes et aux colonne de la grille à résoudre, tandis que la troisième dimension permet d'identifier, par un booléen, les chiffres qui sont valides pour chaque case de la grille.

Au départ, la matrice est entièrement remplie avec la valeur **`True`** indiquant que tout chiffre peut potentiellement se trouver sur chacune des cases de la grille. Puis on va itérer sur les chiffres fournis dans la grille de départ, et mettre à jour la variable **`guess`** entre stockant une valeur **`False`** pour chaque case où le chiffre courant ne pourra être positionné sans générer de conflit (selon les règles du Sudoku, aucun chiffre ne peut être répété sur une même ligne, une même colonne ou un même bloc ***3x3***). De manière duale, s'il n'existe qu'une seule case dans une ligne, une colonne ou un bloc donné, où est affecté une valeur **`True`**, cela signifie que le chiffre correspondant est forcément placé dans cette case, ce qui va interdire de nouvelles cases pour ce chiffre dans d'autres zones de la matrice, et ainsi de suite. L'avantage d'utiliser **numpy** est que toutes les mises à jour de la matrice **`guess`** deviennent triviales grâce aux différents mécanismes disponibles pour l'indexation des matrices.

<h2 style="padding:16px; color:#FFF; background:#07D">Exercice D5 - Ensembles de Julia & Mandelbrot</h2>

Les **ensembles de Julia/Mandelbrot** sont définis en étudiant la convergence/divergence d'une suite à valeurs complexes, processus mathématique appelé ***dynamique holomorphe***. Plus précisément, pour chaque point $z$ du plan complexe <font size=+1>&#8450;</font>, on définit une famille de suites paramétriques :

$$z_0 = z \;\;\;\; \textrm{et} \;\;\;\; z_{k+1} = \,z_k\,\!^p + \,a\,z_0 + \,b$$

où $(a,b)$ sont deux paramètres à valeurs complexes et $p$ un paramètre entier, permettant de faire varier le comportement des suites pour une même valeur $z_0$. Le cas particulier $a = 0$ donne les [**ensembles de Julia**](https://fr.wikipedia.org/wiki/Ensemble_de_Julia) (nommés d'après Gaston Julia, créateur de la dynamique holomorphe) et le cas particulier $(p,a,b) = (2,1,0)$ donne l'[**ensemble de Mandelbrot**](https://fr.wikipedia.org/wiki/Ensemble_de_Mandelbrot) (nommé d'après Benoît Mandelbrot, découvreur des fractales).

On va étudier cette convergence/divergence par un processus de simulation qui commence par échantillonner de manière uniforme, un rectangle $(x_{min}, y_{min}, x_{max}, y_{max})$ pour créer une matrice à valeurs complexes $[z_{ij} = x_{ij} + {\rm i} y_{ij}]$. Puis, pour chaque élément $z_{ij}$, on calcule son **orbite** (c'est-à-dire, les termes successifs de la suite ci-dessus) et on cherche à déterminer le rang $n_{ij}$ pour lequel l'orbite va diverger. Concrètement, il suffit de détecter le nombre d'itérations $k$ avant d'avoir un terme $z_k$ qui vérifie $|z_k| > 2$, car les termes suivants vont alors s'éloigner de plus en plus de l'origine, ce qui implique la divergence de la suite. En appliquant une échelle de couleurs aux différents valeurs $n_{ij}$ calculées, on obtient une visualisation des différentes vitesses de divergence, qui produisent des surfaces fractales, très intéressantes à la fois sur le plan mathématique, mais également sur le plan purement esthétique.
