---
En faisant le calcul des propabilités théoriques ***(c'est un très bon exercice de probabilités que je vous conseille fortement de tenter de résoudre par vous-même)***, on peut trouver la formule analytique pour $P_1$ (probabilité de victoire si on choisit la porte 1) et $P_2$ (probabilité de victoire si on choisit la porte 2) :

$$
P_1 = \frac{m}{n}\qquad {\sf et}\qquad P_2 = \frac{m}{n}\times\frac{n-1}{n-2}
$$

Comme $n\!-\!1$ sera toujours supérieur à $n\!-\!2$, on aura toujours $P_2 > P_1$. Autrement dit, dans toutes les configurations, il sera toujours plus intéressant de changer de porte plutôt que de conserver le choix de la porte initiale.

Le code associé à ces probabilités théoriques est fourni via la fonction **`monty_hall_theoretical`** ci-dessous. Utiliser ce code pour valider la version généralisée de votre implémentation :

````python
def monty_hall_theoretical(n, m):
    """return theoretical winning rates for the generalized (n,m) Monty Hall problem"""
    return m/n, m/n * (n-1)/(n-2) # winning rates for selecting door1 vs door2
````

<h2 style="padding:16px; color:#FFF; background:#06D">Exercice D4 - Manipulation d'images</h2>

Une image numérique est définie comme une matrice de pixels où chaque pixel représente une couleur. Il existe trois grandes familles d'images selon la quantité d'information stockée dans chaque pixel :

- Les **images binaires** où chaque pixel ne contient qu'une valeur binaire 0 (= noir) ou 1 (= blanc)
- Les **images monochromes** où chaque pixel contient un scalaire correspondant à l'intensité lumineuse, qui peut être soit de type **`float`** (sur une échelle de 0 à 1), soit de type **`int`** (sur une échelle de 0 à 255)
- Les **images polychromes** où chaque pixel contient trois scalaires correspondant à l'intensité des trois couleurs primaires (***RGB*** en synthèse additive, ***CMY*** en synthèse soustractive) avec la même échelle que les images monochromes. Certains formats d'images (PNG, par exemple) utilisent une quatrième composante, appelée ***alpha*** et notée ***A***, qui permet de stocker une information de transparence

Dans l'écosystème ***Scientific Python***, on trouve les modules **`PIL`** (ou sa version plus récente, **`Pillow`**) ainsi que **`ndimage`** (fourni par le package **`scipy`**) qui ont été spécialement conçus pour mettre en oeuvre des traitements divers et variés sur des images (images de dimension 2 pour **PIL/Pillow**, images de dimensions arbitraires pour **ndimage**). Lorsqu'on souhaite faire du traitement d'images élaboré, ces modules sont évidemment à utiliser en priorité.

Mais dans le cadre de cet exercice, l'objectif est simplement d'utiliser certaines des fonctionnalités du package **`numpy`** pour tester différentes manipulations sur des images qui auront été converties vers un conteneur de type **`array`**. Une information importante pour analyser une image consiste à tracer l'histogramme des valeurs des pixels, c'est-à-dire compter le nombre de pixels pour chacune des valeurs possibles. Pour une meilleure lisibilité, on génère habituellement les histogrammes par composantes : R/G/B pour les images polychromes, L pour les images monochromes. ### 5 - Tables de correspondance

Une technique classique pour modifier le contenu d'une matrice à valeurs entières consiste à créer et utiliser une table de correspondance (***lookup table*** ou ***LUT*** en abrégé) qui permet d'associer une nouvelle valeur à chaque valeur initiale de la table. Dans le cas d'une image monochrome ou polychrome, les valeurs des pixels sont définies sur l'intervalle $[0, 255]$ donc la table doit posséder 256 éléments, chaque élément de la table étant lui-même défini sur l'intervalle $[0, 255]$.

Par exemple, on peut créer une table contenant 128x la valeur 0 et 128x la valeur 255 :

````python
table = np.repeat([0,255], 128); print(table)
````

Lorsqu'on applique une image monochrome à travers cette table de correspondance, elle est transformée en image binaire (tous les pixels inférieurs à 128 deviennent noirs et tous les pixels supérieurs à 127 deviennent blancs) :

````python
view(table[tigerG]);
````

Lorsqu'on applique la même table sur une une image polychrome, chacune des 3 composantes RGB est binarisée :

````python
view(table[tiger]);
````

Au lieu de créer les tables de correspondance directement, il est plus simple de définir une fonction analytique de [0,1] dans [0,1] et de s'en servir pour créer la table contenant les 256 éléments. La fonction **`view_lut`** ci-dessous, permet de mettre ce principe en oeuvre :

````python
def view_lut(image, function):
  """create lookup table from 'function' and apply to 'image'"""
  ufunc = np.frompyfunc(function, 1, 1) # convert 'function' to ufunc
  ramp = np.linspace(0, 1, 256, endpoint=False) # create a ramp of 256 items in range [0,1[
  table = (255*ufunc(ramp)).astype('u1') # create lookup table
  print(table[::5]) # show subsampling of lookup table (1 sample out of 5)
  view(table[image]) # apply lookup table and show image
````

La table et l'image précédente peuvent ainsi être générées avec une simple fonction **`binary`** qui place un seuil à 0.5 :

````python
binary = lambda t: 0 if t < 0.5 else 1 # la fonction est  toujours définie de [0,1] dans [0,1]
view_lut(tiger, binary) # on obtient le même résultat qu'avec la table construite manuellement
````

Les **ensembles de Julia/Mandelbrot** sont définis en étudiant la convergence/divergence d'une suite à valeurs complexes, processus mathématique appelé ***dynamique holomorphe***. Plus précisément, pour chaque point $z$ du plan complexe <font size=+1>&#8450;</font>, on définit une famille de suites paramétriques :

$$z_0 = z \;\;\;\; {\sf et} \;\;\;\; z_{k+1} = \,z_k\,\!^p + \,a\,z_0 + \,b$$

où $(a,b)$ sont deux paramètres à valeurs complexes et $p$ un paramètre entier, permettant de faire varier le comportement des suites pour une même valeur $z_0$. Le cas particulier $a = 0$ donne les [**ensembles de Julia**](https://fr.wikipedia.org/wiki/Ensemble_de_Julia) (nommés d'après Gaston Julia, créateur de la dynamique holomorphe) et le cas particulier $(p,a,b) = (2,1,0)$ donne l'[**ensemble de Mandelbrot**](https://fr.wikipedia.org/wiki/Ensemble_de_Mandelbrot) (nommé d'après Benoît Mandelbrot, découvreur des fractales).

On va étudier cette convergence/divergence par un processus de simulation qui commence par échantillonner de manière uniforme, un rectangle $(x_{min}, y_{min}, x_{max}, y_{max})$ pour créer une matrice à valeurs complexes $[z_{ij} = x_{ij} + {\sf i} y_{ij}]$. Puis, pour chaque élément $z_{ij}$, on calcule son **orbite** (c'est-à-dire, les termes successifs de la suite ci-dessus) et on cherche à déterminer le rang $n_{ij}$ pour lequel l'orbite va diverger. Concrètement, il suffit de détecter le nombre d'itérations $k$ avant d'avoir un terme $z_k$ qui vérifie $|z_k| > 2$, car les termes suivants vont alors s'éloigner de plus en plus de l'origine, ce qui implique la divergence de la suite. En appliquant une échelle de couleurs aux différents valeurs $n_{ij}$ calculées, on obtient une visualisation des différentes vitesses de divergence, qui produisent des surfaces fractales, très intéressantes à la fois sur le plan mathématique, mais également sur le plan purement esthétique.
