Un des objectifs de l'exercice C5, à réaliser en travail personnel, est de générer le chemin de longueur minimale reliant l'entrée et la sortie d'un labyrinthe. L'algorithme pour obtenir ce résultat optimal (basé sur l'algorithme récursif de Dijkstra) est relativement complexe à implémenter. Il existe néanmoins un algorithme beaucoup plus simple, appelé **méthode de Tes**, qui peut se résumer par la phrase suivante :

> ***Le marcheur choisit une direction possible (droite ou gauche) à la première intersection rencontrée, puis pour toutes les intersections suivantes, il prend la direction opposée à son choix initial***

Dans le cas des labyrinthes parfaits, comme ceux utilisés dans les exercices B5 et C3, cette méthode permet d'arriver à coup sûr vers la sortie, même si elle ne donne que très rarement le chemin optimal.

Du point de vue algorithmique, il est important de réaliser qu'à tout moment durant son parcours, le marcheur sera défini par trois variables : son indice de ligne **i**, son indice de colonne **j** et sa direction **d** qui peut être représentée par un nombre entier de 0 à 3, numéroté dans le sens horaire : ***0 = Nord, 1 = Est, 2 = Sud, 3 = Ouest***. Ainsi ***tourner à droite*** s'implémente en faisant **`d = (d+1) % 4`** et ***tourner à gauche*** s'obtient en faisant **`d = (d-1) % 4`**. De même, pour gérer les déplacements, il est judicieux de définir une liste **`moves`** regroupant un vecteur de translation pour chacune des quatre directions :

```moves = [(-1, 0), (0, 1), (1, 0), (0, -1)] # 0 = North, 1 = East, 2 = South, 3 = West```

<h2 style="padding:16px; color:#FFF; background:#07D">B3 - Tic Tac Toe ★★★</h2>

Le [**TicTacToe**](https://fr.wikipedia.org/wiki/Tic-tac-toe) est la version la plus simple des jeux d'alignements (communément appelés ***morpion***) qui se joue à deux joueurs sur une grille 3x3. La simplicité du jeu permet d'énumérer sans difficulté, toutes les configurations possibles pour une partie, et créer ainsi un adversaire virtuel qui va toujours jouer le coup optimal, et ainsi ne pourra jamais perdre.

La configuration d'une grille de TicTacToe peut se représenter par une chaîne de 9 caractères, composée des trois caractères suivants : **`X`**, **`O`** et **`.`** (ce dernier caractère représentant une case vide). Ainsi, les 12 grilles figurant sur l'image ci-dessous, peuvent être encodées par la liste suivante :
```
['.........' , 'X........' , '.X.......' , '....X....' , 'XO.......' , 'X.O......',
 'X...O....' , 'X....O...' , 'X.......O' , 'OX.......' , '.X.O.....' , '.X..O....']
```
<center></center>

Dans un premier temps, on cherche à énumérer toutes les **configurations uniques et valides** d'une grille de TicTacToe. L'expression "uniques et valides" signifie que l'on ne va pas retenir l'intégralité des grilles possibles mais uniquement celles qui vérifient les propriétés suivantes :

- Le premier joueur est toujours le joueur **`X`**, puis les joueurs alternent leur coup, donc les grilles valides ont forcément, soit le même nombre **`X`** et de **`O`**, soit un nombre de **`X`** supérieur de 1 au nombre de **`O`**
