Le [**jeu de Dames**](https://fr.wikipedia.org/wiki/Dames#Les_variantes_nationales) est un jeu de réflexion sur plateau à deux joueurs, mettant en oeuvre ***n(n-1)*** pions par joueur, ces pions étant répartis sur les cases blanches d'un damier de taille ***2n x 2n***. Le but du jeu est de prendre l'intégralité des pions du joueur adverse, sachant que les règles de prise et de déplacement, varient légèrement selon les variantes. Le plateau de jeu peut se construire pour une valeur quelconque de ***n > 1***, mais les trois variantes les plus connues utilisent les valeurs :

* ***n = 4*** (soit 12 pions par joueur, sur un damier 8x8) appelée **Dames Anglaises**
* ***n = 5*** (soit 20 pions par joueur, sur un damier 10x10) appelée **Dames Françaises**
* ***n = 6*** (soit 30 pions par joueurs, sur un damier 12x12) appelée **Dames Canadiennes**
</ul>
