<h2 style="padding:16px; color:#FFF; background:#07D">C1 - Enumération de tuiles à <i>m</i> sommets et <i>n</i> couleurs ★</h2>

De nombreux algorithmes sur les graphes utilisent le principe de la ***coloration des noeuds et/ou des arêtes***, afin d'établir et de démontrer diverses propriétés associées à certaines familles de graphes. On a pu montrer, dans le domaine de la combinatoire, que beaucoup de ces algorithmes de coloration de graphes étaient équivalents à des algorithmes de ***manipulation de tuiles colorées***, qui sont souvent beaucoup plus simples à implémenter.

Dans sa définition la plus générale, une **tuile colorée** est un polygone régulier quelconque auquel est associé une couleur à chaque sommet (pour une meilleure visualisation, on va généralement colorier toute une zone de la tuile, depuis le sommet jusqu'au barycentre). Concrètement, une tuile colorée va être encodée par une simple chaîne de caractères, en associant une lettre à chaque couleur différente, et en énumérant les couleurs des sommets dans un ordre prédéfini (généralement le sens horaire en partant d'en haut à gauche). Par exemple, sur la figure ci-dessous, on a associé les lettres **A, B, C, D** aux couleurs **rouge, vert, bleu, jaune**. Les trois tuiles peuvent donc s'encoder respectivement par les chaînes **ABCD, DABC** et **ABDC**.

<center></center><br>

Une caractéristique importante des tuiles colorées est que ***deux tuiles sont considérées comme équivalentes si l'une peut être obtenue en appliquant une rotation sur l'autre*** (mathématiquement, cela revient à définir des classes d'équivalence entre tuiles par rapport à l'invariance par rotation). Sur la figure précédente, on constate ainsi que les deux tuiles de gauche sont équivalentes (il suffit d'une rotation de 90° pour passer de l'une à l'autre) alors que la troisième ne pourra être obtenue par rotation à partir des deux premières. Avec l'encodage des tuiles par chaîne de caractères, les classes d'équivalences sont facilement obtenues par permutations circulaires : ***ABCD ~ BCDA ~ CDAB ~ DABC***

Pour une tuile définie par un polygone à ***m*** sommets et coloriée avec un jeu de ***n*** couleurs, il est facile de montrer que le nombre total de configurations possibles est de <b><i>n<sup>m</sup></i></b>. Mais en supprimant toutes les configurations équivalentes, le nombre de configuration restantes est beaucoup plus difficile à déterminer. Il n'existe pas, à ce jour, de formule analytique permettant de donner directement cette valeur en fonction de ***m*** et de ***n***, donc la seule issue possible est d'énumérer ces configurations par algorithme, ce qui est l'objet de cet exercice.
