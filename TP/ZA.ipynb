<h2 style="padding:16px; color:#FFF; background:#07D">A2 - Sortie de labyrinthe par la méthode de la main droite ★★</h2>

Un des objectifs de l'exercice C5, à réaliser en travail personnel, est de générer le chemin de longueur minimale reliant l'entrée et la sortie d'un labyrinthe. L'algorithme pour obtenir ce résultat optimal (basé sur l'algorithme récursif de Dijkstra) est relativement complexe à implémenter. Il existe néanmoins un algorithme beaucoup plus simple, appelé **méthode de la main droite**, qui peut se résumer par la phrase suivante :

> ***Le marcheur entre dans le labyrinthe en touchant, avec sa main, le mur à sa droite, et va suivre uniquement le chemin qui lui permet de ne jamais rompre le contact avec ce mur de droite***

Dans le cas des labyrinthes parfaits, comme ceux utilisés dans les exercices B5 et C3, cette méthode permet d'arriver à coup sûr vers la sortie, même si elle ne donne que très rarement le chemin optimal. Il existe également un algorithme dual, appelé **méthode de la main gauche**, qui fonctionne de manière équivalente en gardant le contact avec le mur de gauche, depuis l'entrée jusqu'à la sortie du labyrinthe.

Du point de vue algorithmique, il est important de réaliser qu'à tout moment durant son parcours, le marcheur sera défini par trois variables : son indice de ligne **i**, son indice de colonne **j** et sa direction **d** qui peut être représentée par un nombre entier de 0 à 3, numéroté dans le sens horaire : ***0 = Nord, 1 = Est, 2 = Sud, 3 = Ouest***. Ainsi ***tourner à droite*** s'implémente en faisant **`d = (d+1) % 4`** et ***tourner à gauche*** s'obtient en faisant **`d = (d-1) % 4`**. De même, pour gérer les déplacements, il est judicieux de définir une liste **`moves`** regroupant un vecteur de translation pour chacune des quatre directions :

```moves = [(-1, 0), (0, 1), (1, 0), (0, -1)] # 0 = North, 1 = East, 2 = South, 3 = West```

On comparant les algorithmes implémentant la méthode de la main droite et celle de la main gauche, on s'aperçoit que les codes sont quasiment identiques. La seule chose qui change est que chaque fois que la méthode de la main droite fait un +1, celle de la main gauche fait un -1, et vice versa. Ce qui signifie qu'on peut implémenter les deux techniques dans une même fonction, en rajoutant simplement un paramètre pour distinguer les deux versions.
